```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
X, y = make_classification(n_samples=200, n_features=2, n_classes=2, n_informative=2,n_redundant=0, random_state=42)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
def plot_decision_boundary(model, X, y, title):
      x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
      y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
      xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),np.arange(y_min, y_max, 0.01))
      Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
      Z = Z.reshape(xx.shape)
      plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.coolwarm)
      plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap=plt.cm.coolwarm)
      plt.title(title)
      plt.show()

```

```python
linear_svm = SVC(kernel='linear', random_state=42)
linear_svm.fit(X_train, y_train)
```

```python
y_pred_linear = linear_svm.predict(X_test)
accuracy_linear = accuracy_score(y_test, y_pred_linear)
conf_matrix_linear = confusion_matrix(y_test, y_pred_linear)
```

```python
print(f"Linear Kernel SVM Accuracy: {accuracy_linear:.2f}")
print("Confusion Matrix (Linear Kernel):")
print(conf_matrix_linear)
ConfusionMatrixDisplay(conf_matrix_linear).plot()
```

```python
plot_decision_boundary(linear_svm, X, y, "Decision Boundary (Linear Kernel)")
```

```python
rbf_svm = SVC(kernel='rbf', random_state=42)
rbf_svm.fit(X_train, y_train)
```

```python
y_pred_rbf = rbf_svm.predict(X_test)
accuracy_rbf = accuracy_score(y_test, y_pred_rbf)
conf_matrix_rbf = confusion_matrix(y_test, y_pred_rbf)
```

```python
print(f"\nRBF Kernel SVM Accuracy: {accuracy_rbf:.2f}")
print("Confusion Matrix (RBF Kernel):")
print(conf_matrix_rbf)
ConfusionMatrixDisplay(conf_matrix_rbf).plot()
```

```python
plot_decision_boundary(rbf_svm, X, y, "Decision Boundary (RBF Kernel)")
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
scaled_svm = SVC(kernel='rbf', random_state=42)
scaled_svm.fit(X_train_scaled, y_train)
```

```python
y_pred_scaled = scaled_svm.predict(X_test_scaled)
accuracy_scaled = accuracy_score(y_test, y_pred_scaled)
conf_matrix_scaled = confusion_matrix(y_test, y_pred_scaled)
```

```python
print(f"\nRBF Kernel SVM with Feature Scaling Accuracy: {accuracy_scaled:.2f}")
```

```python

print("Confusion Matrix (RBF Kernel with Scaling):")
```

```python
print(conf_matrix_scaled)
ConfusionMatrixDisplay(conf_matrix_scaled).plot()
```
